```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 

data = pd.read_csv('data.csv')
n = len(data)
num_plays = data['play_id']
```

```python
print(data['player_x'])
print(data['player_y'])
plt.scatter(data['player_x'], data['player_y'], marker='o', s=1)
plt.show()
```

```python
play_types = data['play_type'].unique().ravel()
play_type_dict = {value: index for index, value in enumerate(play_types)}
num_plays = len(play_types)
subsequent_play = np.zeros((num_plays, num_plays))

current_play_num = 0
for i in range(n):
    current_play = data['play_type'][i]
    if current_play_num == data['play_id'][i]:
        continue
    elif data['play_id'][i] == 1:
        current_play_num = 1
    else:
        previous_play = data['play_type'][i-1]
        current_play_num += data['play_id'][i]-data['play_id'][i-1]
        subsequent_play[play_type_dict[previous_play]][play_type_dict[current_play]] += 1

for i in range(len(subsequent_play)):
    total = subsequent_play[i].sum()
    for j in range(len(subsequent_play[i])):
        
        subsequent_play[i][j] = subsequent_play[i][j] / total * 100
```

```python
play_type_dict = {index: value for index, value in enumerate(play_types)}

for i in range(len(subsequent_play)):
    print(list(play_type_dict.values()), subsequent_play[i])
    plt.barh(list(play_type_dict.values()), subsequent_play[i])
    plt.title(play_type_dict[i])
    plt.xlabel("Percent")
    plt.ylabel("Subsequent play")
    plt.show()

```

```python
result_types = data['result_type'].unique()
num_results = len(result_types)
result_type_dict = {value: index for index, value in enumerate(result_types)}
sequences = {}
current_sequence = []
current_play_id = 0
current_sequence_id = data['sequence_id'][0]
i = 0

while i < n:
    if current_sequence_id == data['sequence_id'][i]:
        if current_play_id == data['play_id'][i]:
            current_sequence.append(data['play_type'][i])
            current_play_id += 1
            i += 1
        elif current_play_id < data['play_id'][i]:
            current_play_id = data['play_id'][i]
        else:
            i += 1
    else:
        sequence = tuple(current_sequence)
        result = data['result_type'][i-1]
        result_pts = data['result_pts'][i-1]
        if  pd.isnull(result_pts):
            result_pts = 0
        if sequence not in sequences:
            sequences[sequence] = [[0, 0] for _ in range(num_results)] + [0, 0, 0]
        sequences[sequence][result_type_dict[result]][0] += 1
        sequences[sequence][-2] += result_pts
        if result_pts > 0:
            sequences[sequence][result_type_dict[result]][1] += 1
            sequences[sequence][-3] += 1
        current_sequence = []
        current_sequence_id = data['sequence_id'][i]
        current_play_id = data['play_id'][i]
        
```

```python
sequence_list = list(sequences.keys())
num_sequences = len(sequence_list)
for i in range(num_sequences):
    total = sum(sublist[0] for sublist in sequences[sequence_list[i]][:-3])
    print(total)
    sequences[sequence_list[i]][-1] = total
    sequences[sequence_list[i]][-2] = sequences[sequence_list[i]][-2] / total
    for j in range(num_results):
        if sequences[sequence_list[i]][j][0] != 0:
            sequences[sequence_list[i]][j][1] = sequences[sequence_list[i]][j][1] / sequences[sequence_list[i]][j][0]
        sequences[sequence_list[i]][j][0] = sequences[sequence_list[i]][j][0] / total * 100

        
print(sequences)
```

```python
sample_threshold = 200

result_type_dict = {index: value for index, value in enumerate(result_types)}
result_type_dict[0] = "N/A"

for i in range(len(sequence_list)):
    if sequences[sequence_list[i]][-1] < sample_threshold:
        continue
    print("Points per sequence: ", sequences[sequence_list[i]][-2])
    print("Number of sequences: ", sequences[sequence_list[i]][-1])
    plt.barh(list(result_type_dict.values()), [sublist[0] for sublist in sequences[sequence_list[i]][:-3]])
    plt.title(sequence_list[i])
    plt.xlabel("Percent")
    plt.ylabel("Result play")
    plt.show()
```

```python
pt_threshold = 0.8

sequence_dict = {}
for i in range(len(sequence_list)):
    sequence_dict[sequence_list[i]] = sequences[sequence_list[i]]
key_seqs = [str(key) for key, value in sequence_dict.items() if value[-1] > sample_threshold and value[-2] > pt_threshold and 'Transition' not in key]
key_seq_pts = [value[-2] for key, value in sequence_dict.items() if value[-1] > sample_threshold and value[-2] > pt_threshold and 'Transition' not in key]
plt.barh(key_seqs, key_seq_pts)
plt.title(f"Sequences with more than {sample_threshold} occurences and over {pt_threshold} pts per sequence")
```

```python
key_seqs = [key for key, value in sequence_dict.items() if value[-1] > sample_threshold and value[-2] > pt_threshold and 'Transition' not in key]
plt.style.use('fivethirtyeight')
for key in key_seqs:
    print("Pts per sequence: ", sequence_dict[key][-2])
    print("Number of sequences: ", sequence_dict[key][-1])
    print("Number of sequences with points scored: ", sequence_dict[key][-3])
    
    fig, ax = plt.subplots()
    #ax.barh(categories, values1, color='blue', label='Values 1')
    #ax.barh(categories, values2, left=values1, color='orange', label='Values 2')
    ax.barh(list(result_type_dict.values()), [sublist[0] * sublist[1] if sublist[0] != 0 else 0 for sublist in sequence_dict[key][:-3]], color = "seagreen")
    ax.barh(list(result_type_dict.values()), [sublist[0] * (1 - sublist[1]) if sublist[0] != 0 else 0 for sublist in sequence_dict[key][:-3]], left=[sublist[0] * sublist[1] if sublist[0] != 0 else 0 for sublist in sequence_dict[key][:-3]], color="maroon")
    plt.title(str(key))
    plt.xlabel("Percent")
    plt.ylabel("Result play")
    plt.show()
    
```

```python
import copy

play_types = data['play_type'].unique()
players = data['player_name'].unique()
player_type = data['player_type'].unique()
player_type_total = data['player_type'].unique() + ' Total'
player_type_toggle = data['player_type'].unique() + ' Toggle'

player_type_aug = np.concatenate((np.concatenate((player_type, player_type_total)), player_type_toggle))
player_type_dict = {type: 0 for type in player_type_aug}
play_type_dict = {value: copy.deepcopy(player_type_dict) for index, value in enumerate(play_types)}
current_play_id = 0
current_sequence_id = data['sequence_id'][0]
i = 0
player_dict = {player: copy.deepcopy(play_type_dict) for player in players}
current_players = []


while i < n:
    if current_sequence_id == data['sequence_id'][i]:
        if current_play_id == data['play_id'][i]:
            if current_sequence == []:
                current_sequence.append(data['play_type'][i])
            player_dict[data['player_name'][i]][data['play_type'][i]][data['player_type'][i] + ' Toggle'] = 1 
            current_players.append((data['player_name'][i], data['play_type'][i]))
            i += 1
        else:
            current_play_id = data['play_id'][i]
            current_sequence.append(data['play_type'][i])
    else:
        sequence = tuple(current_sequence)
        result = data['result_type'][i-1]
        result_pts = data['result_pts'][i-1]
        if  pd.isnull(result_pts):
            result_pts = 0
        for player, play in current_players:
            for pt in player_type:
                if player_dict[player][play][pt + ' Toggle'] == 1 and sequence in key_seqs:
                    player_dict[player][play][pt] += result_pts
                    player_dict[player][play][pt + ' Total'] += 1
                    player_dict[player][play][pt + ' Toggle'] = 0
        current_players = []
        current_sequence = []
        current_sequence_id = data['sequence_id'][i]
        current_play_id = data['play_id'][i]
        
for player in player_dict.keys():
    for play in play_types:
        for pt in player_type:
            if player_dict[player][play][pt + ' Total'] == 0:
                continue
            player_dict[player][play][pt] /= player_dict[player][play][pt + ' Total']
```

```python
off_threshold = 0.8
def_threshold = 0.6
n_threshold = 10

for key in key_seqs:
    print(key)
    offense_players = []
    offense_screen = []
    defense_players = []
    defense_screen = []
    plays = []
    for play in key:
        plays.append(play)
        print(play)
        print("Defense: ")
        for player in players:
            if player_dict[player][play]['BH Def'] < def_threshold and player_dict[player][play]['BH Def Total'] > n_threshold:
                print(player, play, 'BH Def', player_dict[player][play]['BH Def'], player_dict[player][play]['BH Def Total'])
                defense_players.append(player)
            if player_dict[player][play]['SC Def'] < def_threshold and player_dict[player][play]['SC Def Total'] > n_threshold:
                print(player, play, 'SC Def', player_dict[player][play]['SC Def'], player_dict[player][play]['SC Def Total'])
                defense_screen.append(player)
        print("Offense: ")
        for player in players:
            if player_dict[player][play]['Ball-Handler'] > off_threshold and player_dict[player][play]['Ball-Handler Total'] > n_threshold:
                print(player, play, 'Ball-Handler', player_dict[player][play]['Ball-Handler'], player_dict[player][play]['Ball-Handler Total'])
                offense_players.append(player)
            if player_dict[player][play]['Screener'] > off_threshold and player_dict[player][play]['Screener Total'] > n_threshold:
                print(player, play, 'Screener', player_dict[player][play]['Screener'], player_dict[player][play]['Screener Total'])
                offense_screen.append(player)
    fig, ax = plt.subplots()
    for play in key:
        ax.barh(defense_players, [player_dict[player][play]['BH Def'] for player in defense_players], color='firebrick', label='Ball Handler Defense')
        ax.barh(offense_players, [player_dict[player][play]['Ball-Handler'] for player in offense_players], color='yellowgreen', label='Ball Handler')
        ax.barh(defense_screen, [player_dict[player][play]['SC Def'] for player in defense_screen], color='maroon', label='Screen Defense')
        ax.barh(offense_screen, [player_dict[player][play]['Screener'] for player in offense_screen], color='seagreen', label='Screener')
    plt.title(str(key).replace(")", "").replace("(", "").replace("'", ""))
    ax.set_yticklabels([])
    plt.legend()
    plt.xlabel("Average Points")
    plt.show()
 
```
